```python
import pandas as pd
# Load the dataset
data = pd.read_csv('cleaned_data.csv')

# Display the first few rows
data.head()
```

```python
# Clean 'Location' values before dropping duplicates
data['Location'] = data['Location'].astype(str).str.strip().str.lower()

# Now dropna and get unique locations
unique_location_table = (
    data[['Location']]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
)

# Assign new Plant IDs
unique_location_table['Assembly Plant'] = ['Plant ' + str(i+1) for i in range(len(unique_location_table))]

# Save to CSV
unique_location_table.to_csv("Assembly_Plant.csv", index=False)
```
